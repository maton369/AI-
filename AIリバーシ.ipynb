```python
import numpy as np
import random
import os
from PIL import Image
from matplotlib import pylab as plt
import shutil
import glob

# グローバル変数の初期化
global board, gboard
global aip_x, aip_y, man_x, man_y
global aip1_hyouka

# 定数の定義
NONE = 0
BLACK = 1
WHITE = 2

A = 1
B = 2
C = 3
D = 4
E = 5
F = 6
G = 7
H = 8

# 方向の定義 (上下左右および対角線方向)
shui_x = np.array([0, -1, 0, 1, -1, 1, -1, 0, 1])  # x方向のオフセット
shui_y = np.array([0, -1, -1, -1, 0, 0, 1, 1, 1])  # y方向のオフセット

def disp_board():
    """ボードをコンソールに表示する関数"""
    print('    A  B  C  D  E  F  G  H')
    for y in range(1, 9):
        print(y, ' ', end="")
        for x in range(1, 9):
            if board[y, x] == NONE:
                print('―', end="")
            elif board[y, x] == WHITE:
                print('〇', end="")
            elif board[y, x] == BLACK:
                print('●', end="")
            print(' ', end="")
        print('\n', end="")

def init_board():
    """ボードを初期化する関数"""
    global board
    board = np.zeros([10, 10])
    board[4, 4] = WHITE
    board[5, 5] = WHITE
    board[5, 4] = BLACK
    board[4, 5] = BLACK

def init_board2():
    """ボードを別の初期状態に設定する関数"""
    global board
    board = np.ones([10, 10])
    board[4, 4] = WHITE
    board[5, 5] = WHITE
    board[5, 4] = BLACK
    board[4, 5] = BLACK
    board[1, 1] = 0
    board[1, 2] = 0
    board[1, 3] = 0
    board[1, 4] = 0
    board[2, 1] = 0
    board[3, 1] = 0
    board[4, 1] = 0

def okeruka_sub(y, x, turn, shui):
    """指定された位置で石を置けるか確認するサブ関数"""
    okeru_flag = 0
    koma1 = BLACK if turn == BLACK else WHITE
    koma2 = WHITE if turn == BLACK else BLACK

    if board[y, x] != NONE:
        return 0

    temp_x = x + shui_x[shui]
    temp_y = y + shui_y[shui]

    if board[temp_y, temp_x] == NONE:
        return 0
    elif board[temp_y, temp_x] == koma1:
        return 0
    elif board[temp_y, temp_x] == koma2:
        while True:
            temp_x = temp_x + shui_x[shui]
            temp_y = temp_y + shui_y[shui]

            if temp_x <= 0 or temp_y <= 0 or temp_x >= 9 or temp_y >= 9:
                return 0

            if board[temp_y, temp_x] == koma2:
                continue
            elif board[temp_y, temp_x] == koma1:
                return 1
            elif board[temp_y, temp_x] == NONE:
                return 0

def okeruka(oku_y, oku_x, turn):
    """指定された位置に石を置けるか確認する関数"""
    for i in range(1, 9):
        oku = okeruka_sub(oku_y, oku_x, turn, i)
        if oku == 1:
            return oku
    return 0

def uragaesu_sub(y, x, turn, shui):
    """指定された位置に石を裏返すサブ関数"""
    global board

    koma1 = BLACK if turn == BLACK else WHITE
    koma2 = WHITE if turn == BLACK else BLACK

    temp_x = x + shui_x[shui]
    temp_y = y + shui_y[shui]

    while True:
        if board[temp_y, temp_x] == koma1:
            return 0
        elif board[temp_y, temp_x] == koma2:
            board[temp_y, temp_x] = koma1
        temp_x = temp_x + shui_x[shui]
        temp_y = temp_y + shui_y[shui]

def uragaesu(oku_y, oku_x, turn):
    """指定された位置に石を置いたときに裏返る石を裏返す関数"""
    global board
    okeru = 0
    koma1 = BLACK if turn == BLACK else WHITE
    koma2 = WHITE if turn == BLACK else BLACK

    for i in range(1, 9):
        okeru_flag = okeruka_sub(oku_y, oku_x, turn, i)
        if okeru_flag == 1:
            uragaesu_sub(oku_y, oku_x, turn, i)
            okeru = 1

    if okeru == 1:
        board[oku_y, oku_x] = koma1

    return okeru

def owarika(turn):
    """ゲームの終了をチェックする関数"""
    for y in range(1, 9):
        for x in range(1, 9):
            if okeruka(y, x, turn) == 1:
                return 0

    turn = WHITE if turn == BLACK else BLACK

    for y in range(1, 9):
        for x in range(1, 9):
            if okeruka(y, x, turn) == 1:
                return 1

    return 2

def hantei(taisen):
    """ゲームの結果を判定する関数"""
    bcount = 0
    wcount = 0

    if taisen == 1:
        kuro = 'あなた●'
        siro = 'あいぴ〇'
    elif taisen == 2:
        kuro = 'あいぴ●'
        siro = 'あいぴ〇'

    for y in range(1, 9):
        for x in range(1, 9):
            if board[y, x] == BLACK:
                bcount += 1
            elif board[y, x] == WHITE:
                wcount += 1

    print('＊＊ ●%d：〇%dの結果 ' % (bcount, wcount), end="")
    if bcount > wcount:
        print('%sの勝ち ＊＊' % (kuro), end="")
        return 1
    elif wcount > bcount:
        print('%sの勝ち ＊＊' % (siro), end="")
        return 2
    elif wcount == bcount:
        print('引き分け ＊＊', end="")
        return 3

def aip_random(turn):
    """AIがランダムに石を置く位置を決定する関数"""
    global aip_x, aip_y
    sumi_x = np.array([1, 8, 1, 8])
    sumi_y = np.array([1, 1, 8, 8])

    # 角に置けるか確認し、置ける場合はその位置を選ぶ
    for i in range(0, 4):
        if okeruka(sumi_y[i], sumi_x[i], turn) == 1:
            aip_x = sumi_x[i]
            aip_y = sumi_y[i]
            return 1

    # ランダムに位置を選ぶ
    while True:
        x = random.randint(1, 8)
        y = random.randint(1, 8)
        if okeruka(y, x, turn) == 1:
            aip_x = x
            aip_y = y
            return 1

def aip_othello(i):
    """AI対AIまたはAI対人間のゲームを実行する関数"""
    man_aip = 1  # 人間対あいぴ1
    aip_aip = 2  # あいぴ１対あいぴ２
    taisen = aip_aip  # ゲームの種類

    turn = BLACK  # 黒が先攻
    okeru_flag = 0
    oku_flag = 0
    oku_x = 0
    oku_y = 0

    init_board()

    while True:
        bcount = np.count_nonzero(board)  # 現在の盤面の石の数
        input_flag = 1
        while input_flag == 1:
            owari_flag = owarika(turn)
            if owari_flag == 2:
                kekka = hantei(taisen)  # ゲームの結果を判定
                return kekka

            if owari_flag == 0:
                dummy = 0
            elif owari_flag == 1:
                turn = WHITE if turn == BLACK else BLACK

            if taisen == aip_aip:
                # AIのターン
                if turn == BLACK:
                    aip_random(turn)
                    oku_x = aip_x
                    oku_y = aip_y
                    input_flag = 0
                elif turn == WHITE:
                    aip_random(turn)
                    oku_x = aip_x
                    oku_y = aip_y
                    input_flag = 0

        okeru = okeruka(oku_y, oku_x, turn)
        if okeru == 1 and turn == BLACK:
            disp_gboard(i, bcount, oku_y, oku_x)  # 盤面を画像として保存

        if okeru == 1:
            uragaesu(oku_y, oku_x, turn)  # 石を裏返す
            turn = WHITE if turn == BLACK else BLACK  # プレイヤーを切り替える

def init_gboard():
    """盤面を画像として初期化する関数"""
    global gboard

    dst_img = Image.fromarray(board)
    if dst_img.mode != 'RGB':
        dst_img = dst_img.convert('RGB')
    dst_img.save('board_rgb_in.bmp')

    gboard = np.array(Image.open('board_rgb_in.bmp'))
    gboard[:, :, (0, 1, 2)] = 50  # 盤面の初期表示を設定
    plt.imshow(gboard)
    plt.show()

def disp_gboard(i, no, yy, xx):
    """盤面の状態を画像として表示し、保存する関数"""
    global gboard
    gboard[:, :, (0, 1, 2)] = 50

    for y in range(0, 10):
        for x in range(0, 10):
            if board[y, x] == 1:
                gboard[y, x, 0] = 255  # 黒石の表示
            if board[y, x] == 2:
                gboard[y, x, 1] = 255  # 白石の表示

    filename = f'dload_tmp/ban_{i}_{no-3}_{yy}_{xx}.bmp'
    gboard_save = Image.fromarray(gboard)
    gboard_save.save(filename, 'bmp')

# メイン処理
if __name__ == "__main__":
    # 一時ファイル用ディレクトリの作成
    if not os.path.exists('dload_tmp'):
        os.makedirs('dload_tmp')

    if not os.path.exists('dload'):
        os.makedirs('dload')

    init_board()  # ボードの初期化
    init_gboard()  # 画像ボードの初期化

    # AI対AIのゲームを100回実行
    for i in range(0, 100):
        kekka = aip_othello(i)  # ゲームの実行
        plt.imshow(gboard)
        plt.show()

        if kekka == 1 or kekka == 3:
            print(f'\n＊＊＊ {i}回目:先攻黒あいぴの勝ち ＊＊＊')
            print("＊＊＊ 結果をコピー保存します ＊＊＊")
            disp_gboard(i, 63, aip_y, aip_x)  # 画像保存
            dest_dir = "dload"
            for file in glob.glob('dload_tmp/*.bmp'):
                shutil.copy(file, dest_dir)  # ファイルを保存先にコピー

            for file in glob.glob('dload_tmp/*.bmp'):
                os.remove(file)  # 一時ファイルを削除
        else:
            for file in glob.glob('dload_tmp/*.bmp'):
                os.remove(file)  # 一時ファイルを削除

        print("\n\n")

    print("＊＊＊ データ収集終了 ＊＊＊")
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D, Dropout
from keras.utils import to_categorical
from keras.preprocessing.image import img_to_array, load_img
import re
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

def list_pictures(directory, ext='jpg|jpeg|bmp|png|ppm'):
    return [os.path.join(root, f)
            for root, _, files in os.walk(directory) for f in files
            if re.match(r'([\w]+\.(?:' + ext + '))', f.lower())]

# main
print("＊＊＊ メインルーチンスタート ＊＊＊")

x = []
y = []

print("＊＊＊ 画像ファイル処理開始 ＊＊＊")

for picture in list_pictures('dload'):
    img_rgb = np.array(Image.open(picture).convert('RGB'))
    filename = os.path.splitext(os.path.basename(picture))[0]
    filename = filename.split('_')
    ys = int(filename[-2])
    xs = int(filename[-1])
    oku_no = (ys - 1) * 8 + xs
    x.append(img_rgb)
    y.append(oku_no)

x = np.array(x)
y = np.array(y)

np.save("20221001_aip_rgb_x_test", x.astype('float32'))
np.save("20221001_aip_rgb_y_test", y.astype('int32'))

x = np.load('20221001_aip_rgb_x_test.npy')
y = np.load('20221001_aip_rgb_y_test.npy')

y = y.astype('int32')
y = to_categorical(y)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

x_train = x_train.reshape(x_train.shape[0], 10, 10, 3)
x_test = x_test.reshape(x_test.shape[0], 10, 10, 3)

# CNNモデルの構築
model_NN = Sequential()
model_NN.add(Conv2D(64, (3, 3), activation="relu", input_shape=(10, 10, 3)))
model_NN.add(Conv2D(32, (3, 3), activation="relu"))
model_NN.add(MaxPooling2D(pool_size=(2, 2)))
model_NN.add(Dropout(rate=0.3))
model_NN.add(Flatten())
model_NN.add(Dense(128, activation='relu'))
model_NN.add(Dense(65, activation='softmax'))

model_NN.summary()

model_NN.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("x_train_shape=", x_train.shape)
print("y_train_shape=", y_train.shape)

print("fit=")
my_epochs = 100

result = model_NN.fit(x_train, y_train, epochs=my_epochs, validation_split=0.2)

# モデルの保存
print("＊＊＊ ニューラルネットワークと重みを保存しました ＊＊＊")
model_NN.save("aip_rgb_model.hdf5")

# 訓練過程の可視化
plt.plot(range(1, my_epochs + 1), result.history['accuracy'], label="training")
plt.plot(range(1, my_epochs + 1), result.history['val_accuracy'], label="validation")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from keras.models import load_model

# グローバル変数の定義
global board, gboard
global aip_x, aip_y, man_x, man_y
global ok_count

NONE = 0
BLACK = 1
WHITE = 2

shui_x = np.array([0, -1, 0, 1, -1, 1, -1, 0, 1])  # 方向ベクトル
shui_y = np.array([0, -1, -1, -1, 0, 0, 1, 1, 1])  # 方向ベクトル

def disp_board():
    print('    A  B  C  D  E  F  G  H')
    for y in range(1, 9):
        print(y, ' ', end="")
        for x in range(1, 9):
            if board[y, x] == NONE:
                print('―', end="")
            elif board[y, x] == WHITE:
                print('〇', end="")
            elif board[y, x] == BLACK:
                print('●', end="")
            print(' ', end="")
        print('\n', end="")

def init_board():
    global board
    board = np.zeros([10, 10])
    board[4, 4] = WHITE
    board[5, 5] = WHITE
    board[5, 4] = BLACK
    board[4, 5] = BLACK

def okeruka_sub(y, x, turn, shui):
    if turn == BLACK:
        koma1 = BLACK
        koma2 = WHITE
    else:
        koma1 = WHITE
        koma2 = BLACK

    if board[y, x] != NONE:
        return 0

    temp_x = x + shui_x[shui]
    temp_y = y + shui_y[shui]

    if board[temp_y, temp_x] == NONE:
        return 0
    elif board[temp_y, temp_x] == koma1:
        return 0
    elif board[temp_y, temp_x] == koma2:
        while True:
            temp_x += shui_x[shui]
            temp_y += shui_y[shui]
            if temp_x <= 0 or temp_y <= 0 or temp_x >= 9 or temp_y >= 9:
                return 0
            if board[temp_y, temp_x] == koma2:
                continue
            elif board[temp_y, temp_x] == koma1:
                return 1
            elif board[temp_y, temp_x] == NONE:
                return 0

def okeruka(oku_y, oku_x, turn):
    for i in range(1, 9):
        if okeruka_sub(oku_y, oku_x, turn, i) == 1:
            return 1
    return 0

def uragaesu_sub(y, x, turn, shui):
    if turn == BLACK:
        koma1 = BLACK
        koma2 = WHITE
    else:
        koma1 = WHITE
        koma2 = BLACK

    temp_x = x + shui_x[shui]
    temp_y = y + shui_y[shui]

    while True:
        if board[temp_y, temp_x] == koma1:
            return
        elif board[temp_y, temp_x] == koma2:
            board[temp_y, temp_x] = koma1
        temp_x += shui_x[shui]
        temp_y += shui_y[shui]

def uragaesu(oku_y, oku_x, turn):
    if turn == BLACK:
        koma1 = BLACK
        koma2 = WHITE
    else:
        koma1 = WHITE
        koma2 = BLACK

    okeru = 0
    for i in range(1, 9):
        if okeruka_sub(oku_y, oku_x, turn, i) == 1:
            uragaesu_sub(oku_y, oku_x, turn, i)
            okeru = 1

    if okeru == 1:
        board[oku_y, oku_x] = koma1
    return okeru

def owarika(turn):
    for y in range(1, 9):
        for x in range(1, 9):
            if okeruka(y, x, turn) == 1:
                return 0
    if turn == BLACK:
        turn = WHITE
    else:
        turn = BLACK

    for y in range(1, 9):
        for x in range(1, 9):
            if okeruka(y, x, turn) == 1:
                return 1

    return 2

def hantei(taisen):
    bcount = 0
    wcount = 0

    if taisen == 1:
        kuro = 'あなた●'
        siro = 'あいぴ〇'
    elif taisen == 2:
        kuro = 'あいぴ●'
        siro = 'あいぴ〇'

    for y in range(1, 9):
        for x in range(1, 9):
            if board[y, x] == BLACK:
                bcount += 1
            elif board[y, x] == WHITE:
                wcount += 1

    print('\n＊＊ ●%d：〇%dの結果 ' % (bcount, wcount), end="")
    if bcount > wcount:
        print('%sの勝ち ＊＊' % (kuro), end="")
        return 1
    elif wcount > bcount:
        print('%sの勝ち ＊＊' % (siro), end="")
        return 2
    else:
        print('引き分け ＊＊', end="")
        return 3

def man(turn):
    global man_x, man_y
    print(' a1 or A1> ', end="")
    inp_ab = input()
    a, b = inp_ab[:1], inp_ab[1:]
    if ord(a) >= 0x61:  # 小文字
        man_x = ord(a) - 0x60
    else:  # 大文字
        man_x = ord(a) - 0x40
    man_y = int(b)

def aip_random(turn):
    global aip_x, aip_y
    sumi_x = np.array([1, 8, 1, 8])
    sumi_y = np.array([1, 1, 8, 8])

    for i in range(4):
        if okeruka(sumi_y[i], sumi_x[i], turn) == 1:
            aip_x = sumi_x[i]
            aip_y = sumi_y[i]
            return 1

    while True:
        x = random.randint(1, 8)
        y = random.randint(1, 8)
        if okeruka(y, x, turn) == 1:
            print(' 乱数あいぴ > %s%d \n' % (chr(x + 0x60), y))
            aip_x = x
            aip_y = y
            return 1

def aip_deep(turn):
    global aip_x, aip_y
    global ok_count

    henkan = [0, 11, 12, 13, 14, 15, 16, 17, 18,
              21, 22, 23, 24, 25, 26, 27, 28,
              31, 32, 33, 34, 35, 36, 37, 38,
              41, 42, 43, 44, 45, 46, 47, 48,
              51, 52, 53, 54, 55, 56, 57, 58,
              61, 62, 63, 64, 65, 66, 67, 68,
              71, 72, 73, 74, 75, 76, 77, 78,
              81, 82, 83, 84, 85, 86, 87, 88]

    x_img = np.array(Image.open('ban_now.bmp').convert('RGB'))
    x = [x_img]
    x = np.array(x)
    x = x.astype('float32') / 255.0
    x_target = x.reshape(x.shape[0], 10, 10, 3)

    # 学習済みの重みのロード
    model_NN = load_model("aip_rgb_model.hdf5")

    while True:
        y_prob = model_NN.predict(x_target, batch_size=1)
        y_prob = y_prob.reshape(64)
        y_prob = np.array(y_prob)

        idx = np.argmax(y_prob)
        if y_prob[idx] > 0.01 and idx in henkan:
            aip_x = (idx % 8) + 1
            aip_y = (idx // 8) + 1
            if okeruka(aip_y, aip_x, turn) == 1:
                print(' 深層あいぴ > %s%d \n' % (chr(aip_x + 0x60), aip_y))
                return 1
            else:
                y_prob[idx] = 0.0
                continue

        print(' 不正確な結果で、別の場所を試します。')
        return 0

def aip_riversi(turn, ai):
    global ok_count
    ok_count = 0
    if ai == 1:
        aip_random(turn)
    elif ai == 2:
        aip_deep(turn)

def p_turn(turn, ai):
    global ok_count
    print('   #', end="")
    if turn == BLACK:
        print(' 黒のターン', end="")
    else:
        print(' 白のターン', end="")
    print('\n', end="")

    if turn == BLACK:
        if ai == 1:
            aip_riversi(turn, 1)
            if ok_count == 0:
                print(' スキップします\n', end="")
                turn = WHITE
                return turn
        elif ai == 2:
            aip_riversi(turn, 2)
            if ok_count == 0:
                print(' スキップします\n', end="")
                turn = WHITE
                return turn

        man(turn)
        if okeruka(man_y, man_x, turn) == 1:
            uragaesu(man_y, man_x, turn)
            ok_count = 1
        else:
            print(' 不正解です。別の場所を試します。')
            ok_count = 0
            turn = WHITE

    else:
        if ai == 1:
            aip_riversi(turn, 1)
            if ok_count == 0:
                print(' スキップします\n', end="")
                turn = BLACK
                return turn
        elif ai == 2:
            aip_riversi(turn, 2)
            if ok_count == 0:
                print(' スキップします\n', end="")
                turn = BLACK
                return turn

        man(turn)
        if okeruka(man_y, man_x, turn) == 1:
            uragaesu(man_y, man_x, turn)
            ok_count = 1
        else:
            print(' 不正解です。別の場所を試します。')
            ok_count = 0
            turn = BLACK

    return turn

def main():
    global board
    global man_x, man_y
    global aip_x, aip_y
    global ok_count

    print('【リバーシゲーム】\n')
    print('対戦する AI の種類を選択してください。')
    print('1: ランダム AI')
    print('2: 学習済みモデル AI')
    ai = int(input('番号を入力してください: '))

    init_board()
    disp_board()

    turn = BLACK
    while True:
        if owarika(turn) == 2:
            break
        turn = p_turn(turn, ai)
        disp_board()

    hantei(ai)
    print('終了しました。')

if __name__ == "__main__":
    main()
```
